```python
for i in range(len(df['Bộ VXL'])):
    if type(df['Bộ VXL'][i]) != str:
        continue
    if len(df['Bộ VXL'][i]) <= 2:
        df['Bộ VXL'][i] = df['Công nghệ CPU'][i]
```

```python
df_new = pd.DataFrame(columns=[
 'name',
 'brand',
 'chipset',
 'chipset_gen',
 'ram', 
 'ram_max', 
 'vga',
 'storage_type',
 'storage', 
 'os',
 'screen_size',
 'webcam',
 'weight', 
 'price'
])
```

```python
df_new['name'] = df['name']
```

```python
df_new['brand'] = df['name'].str.extract(r'(Asus|Dell|Lenovo|HP|Acer|Apple|Acer|MSI|VAIO|LG|DELL|Chuwi|Lenovo|Microsoft|LENOVO|LENOVO|Dell|ASUS)')
```

```python
df_new['chipset'] = df['Bộ VXL']
df_new['chipset_gen'] = df['Bộ VXL']
```

```python
from ultil import get_first_word, get_cpu_gen
```

```python
df_new['chipset'] = df_new['chipset'].apply(get_first_word)
df_new['chipset_gen'] = df_new['chipset_gen'].apply(get_cpu_gen)
```

```python
df_new['chipset'].value_counts()
```

```python
for i in range(len(df_new['chipset'])):
    if df_new['chipset'][i] == 'Ultra':
        df_new['chipset'][i] = 'Intel'
        df_new['chipset_gen'][i] = 'Ultra ' + df_new['chipset_gen'][i]
```

```python
df_new['chipset_gen'].value_counts()
```

```python
from ultil import get_first_number
```

```python
df_new['ram'] = df['Bộ nhớ RAM']
```

```python
df_new['ram'] = df_new['ram'].apply(get_first_number)
```

```python
df_new['ram'].isna().sum()
```

```python
df_new['ram_max'] = df['Hỗ trợ RAM tối đa']
df_new['ram_max'] = df_new['ram_max'].apply(get_first_number)
```

```python
df_new['os'] = df['Hệ điều hành'].str.extract(r'(Window.\d\d|Window. \d\d|Win \d\d|Dos|Ubuntu|Mac OS|No OS|Fedora|NoOS)')
```

```python
df_new['os']
```

```python
df_new['vga'] = df['Card đồ họa'].str.lower().str.extract(r'(intel|amd|nvidia)')
```

```python
df_new['storage_type'] = df['Loại ổ cứng'].str.extract(r'(SSD|HDD)')
df_new['storage'] = df['Dung lượng ổ cứng'].str.extract(r'(\d* *GB|\d* *TB)')
```

```python
size_info = df_new['storage'].str.extract(r'(\d+)TB|(\d+)GB')
```

```python
df_new['storage'] = size_info.apply(lambda row: int(row[0]) * 1024 if pd.notna(row[0]) else (int(row[1]) if pd.notna(row[1]) else 512), axis=1)
```

```python
df_new['screen_size'] = df['Kích thước màn hình']
df_new['screen_size'] = df_new['screen_size'].apply(get_first_number)
```

```python
df_new['screen_size']
```

```python
df_new['webcam'] = df['Webcam'].notnull().astype(int)
```

```python
from ultil import get_number
```

```python
df_new['weight'] = df['Trọng lượng']
df_new['weight'] = df_new['weight'].apply(get_number)
```

```python
df_new['price'] = df['gia_niem_yet'].str.replace("₫", "")
df_new['price'] = df_new['price'].str.replace(".", "")
```

```python
df_new['screen_revolution'] = df['Kích thước màn hình'].str.extract(r'(\d+(?:\.\d+)?K|FHD|WUXGA|QHD|Full HD)')
```

```python
df_new['screen_revolution'].value_counts()
```

```python
df_new['screen_revolution'][df_new['screen_revolution'].isin(['Full HD'])] = 'FHD'
```

```python
df_new
```

```python
for column in df_new.columns:
    most_frequent_value = df_new[column].mode()[0]
    df_new[column].fillna(most_frequent_value, inplace=True)
```

```python
df_new.to_csv("phucanh.csv", encoding="utf-8-sig")
```
