```python
import pandas as pd
```

```python
hacom = pd.read_csv('hacom.csv')
```

```python
hacom.head()
```

```python
ltw_vp = pd.read_csv('laptopworld_vp.csv')
```

```python
ltw_gm = pd.read_csv('laptopworld_gaming.csv')
```

```python
ltw = pd.concat([ltw_vp, ltw_gm], ignore_index=True)
```

```python
ltw.head()
```

```python
ltw.drop(columns=['screen_revolution'], inplace=True)
```

```python
ltw.rename(columns={'screen_resolution': 'screen_revolution'}, inplace=True)
```

```python
ltw.head()
```

```python
phucanh = pd.read_csv('phucanh.csv')
```

```python
phucanh.head()
```

```python
final = pd.concat([hacom, ltw, phucanh], ignore_index=True)
```

```python
final.columns
```

```python
final.drop(columns=['battery', 'screen_pixels', 'screen_width', 'screen_height', 'screen_full'], inplace=True)
```

```python
final.columns
```

```python
final.to_csv("final.csv", encoding="utf-8-sig")
```
